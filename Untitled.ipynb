# Task1

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set seaborn style for better visuals
sns.set(style="whitegrid")
```

```python
# Load the dataset
try:
    df = pd.read_csv("student-mat.csv", sep=';')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: File not found.")
```

```python
# Display the first 5 rows
df.head()
```

```python
# Check for structure, data types and missing values
df.info()
df.isnull().sum()
```

```python
# Clean data - drop rows with missing values (if any)
df_cleaned = df.dropna()
print(f"Rows after dropping missing values: {df_cleaned.shape[0]}")
```

# Task 2: Basic Data Analysis

```python
# Summary statistics for numeric columns
df_cleaned.describe()
```

```python
# Grouping by 'sex' to see performance based on gender
grouped_by_sex = df_cleaned.groupby('sex')[['G1', 'G2', 'G3']].mean()
grouped_by_sex
```

```python
# Grouping by 'studytime' and calculating mean final grade
grouped_by_study = df_cleaned.groupby('studytime')['G3'].mean()
grouped_by_study
```

# Task 3: Visualizations

```python
# Line Chart - G3 (final grade) by student index to see performance trend

plt.figure(figsize=(10,5))
plt.plot(df_cleaned.index, df_cleaned['G3'], color='purple')
plt.title("Final Grade Trend of Students")
plt.xlabel("Student Index")
plt.ylabel("Final Grade (G3)")
plt.grid(True)
plt.show()
```

```python
# Bar Chart - Average Final Grade by Study Time

plt.figure(figsize=(8,6))
sns.barplot(x=grouped_by_study.index, y=grouped_by_study.values, palette='Blues_d')
plt.title("Average Final Grade by Study Time")
plt.xlabel("Study Time (hours per week)")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

```python
# Histogram - Distribution of Final Grades

plt.figure(figsize=(8,6))
plt.hist(df_cleaned['G3'], bins=10, color='teal', edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Scatter Plot - Relationship between Absences and Final Grade

plt.figure(figsize=(8,6))
sns.scatterplot(x='absences', y='G3', data=df_cleaned, hue='sex')
plt.title("Absences vs Final Grade")
plt.xlabel("Number of Absences")
plt.ylabel("Final Grade (G3)")
plt.legend(title="Sex")
plt.show()
```
